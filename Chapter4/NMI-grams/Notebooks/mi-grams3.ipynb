```python
import pandas as pd
print(pd.__version__)
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings(action='once')

large = 22; med = 15; small = 12
params = {'axes.titlesize': large,
          'legend.fontsize': med,
          'figure.figsize': (16, 10),
          'axes.labelsize': med,
          'axes.titlesize': med,
          'xtick.labelsize': med,
          'ytick.labelsize': med,
          'figure.titlesize': large}
plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style("white")
%matplotlib inline

amount = 10
```

# 1. Composition

```python
sa = [32000]

nmi_dic = {}
for s  in sa:
  name = "df_mi_comp_25meV_{}".format(s)
  nmi_dic[str(s)] = pd.read_pickle(name)['Input Data|e_form']
  nmi_dic[str(s)].replace([np.inf, -np.inf, np.nan], 0 ,inplace=True)
  nmi_dic[str(s)].apply(pd.to_numeric)
  nmi_dic[str(s)] = nmi_dic[str(s)].sort_values(ascending=False)
```

```python
fig, ax = plt.subplots(figsize=(5,4))
b = sns.barplot(y = nmi_dic['32000'].index[1:amount+1],x=(nmi_dic['32000']).values[1:amount+1],ax=ax,palette="Blues_d")
ax.set_xlim([0,0.165])
fig.savefig("target/comp-eform.pdf", bbox_inches = 'tight')
```

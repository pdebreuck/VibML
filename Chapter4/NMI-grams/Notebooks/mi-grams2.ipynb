```python
import pandas as pd
print(pd.__version__)
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings(action='once')

large = 22; med = 14; small = 12
params = {'axes.titlesize': large,
          'legend.fontsize': med,
          'figure.figsize': (16, 10),
          'axes.labelsize': med,
          'axes.titlesize': med,
          'xtick.labelsize': med,
          'ytick.labelsize': med,
          'figure.titlesize': large}
plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style("white")
%matplotlib inline
```

# 1. Composition

```python
sa = [200,400,800,1245]

nmi_dic = {}
for s  in sa:
  name = "df_mi_comp_phonon_{}".format(s)
  nmi_dic[str(s)] = pd.read_pickle(name)['S_300_atom']
  nmi_dic[str(s)].replace([np.inf, -np.inf, np.nan], 0 ,inplace=True)
  nmi_dic[str(s)].apply(pd.to_numeric)
  nmi_dic[str(s)] = nmi_dic[str(s)].sort_values(ascending=False)

amount = 20
```

# 1.1 MI-gram and convergence

```python
fig, [ax1,ax2] = plt.subplots(1,2,figsize=(12,7),sharey=True)

b = sns.barplot(y = nmi_dic['1245'].index[1:amount+1],x=(nmi_dic['1245']).values[1:amount+1],ax=ax1,palette="Blues_d")
ax1.set_xlim([0,0.11])

b = sns.barplot(y = nmi_dic['1245'].index[1:amount+1],x=(nmi_dic['1245']-nmi_dic['800']).values[1:amount+1],ax=ax2,
                palette="Blues_d")
ax2.set_xlim([-0.05,0.05])

fig.savefig("target/comp-phonon.pdf", bbox_inches = 'tight')
#print((nmi_dic['1245']).index[100:200])
```

```python
fig, [ax1,ax2] = plt.subplots(1,2,figsize=(12,7),sharey=True)

b = sns.barplot(y = nmi_dic['1245'].index[-80:-60],x=(nmi_dic['1245']).values[-80:-60],ax=ax1,palette="Blues_d")
ax1.set_xlim([0,0.02])

b = sns.barplot(y = nmi_dic['1245'].index[-80:-60],x=(nmi_dic['1245']-nmi_dic['800']).values[-80:-60],ax=ax2,palette="Blues_d")
ax2.set_xlim([-0.01,0.01])

fig.savefig("target/comp-phonon_least.pdf", bbox_inches = 'tight')
```

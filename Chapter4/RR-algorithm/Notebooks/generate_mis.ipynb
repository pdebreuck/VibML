```python
from sklearn.feature_selection import mutual_info_regression
import pandas as pd
import numpy as np

def nmi_phonon(df,col='S_300_atom'):

    mi_df = pd.DataFrame([],columns=[col],index=df.columns)

    mi_df.loc[:,col] = (mutual_info_regression(df,df[col]))
    S_mi = mutual_info_regression(df[col].values.reshape(-1, 1),df[col])[0]

    diag={}
    for x in df.columns:
            diag[x]=(mutual_info_regression(df[x].values.reshape(-1, 1),df[x]))[0]
            if diag[x] == 0:
                print("WARNING: {} is zero".format(x),flush=True)

    for x in mi_df.index:
        mi_df.loc[x,col] = mi_df.loc[x,col] / ((S_mi + diag[x])/2)


    return mi_df
```
